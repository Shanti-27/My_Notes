```python
# Working with date table
# converting to datetime datatype instead of object
date['date'] = pd.to_datetime(date['date'])
```

```python
date.head()
```

```python
date['Month_Name'] = date['date'].dt.month_name()
```

```python
date.head(10)
```

```python
# part 3 - extract days

date['date_day'] = date['date'].dt.day
```

```python
date.head()
```

```python
# day of week
date['date_dow'] = date['date'].dt.dayofweek
```

```python
date.head()
```

```python
# days name
date['days_name'] = date['date'].dt.day_name()
date.head(10)
```

```python
date['date_dow'].value_counts()
```

```python
date['days_name'].value_counts()
```

```python
date.columns
```

```python
date.drop(columns=['product_id', 'city_id', 'orders'], inplace = True)
```

```python
date.head()
```

```python
np.where()
```

```python
# weekend means 1 and weekday means 0
# is weekend ?
date['date_is_weekend'] = np.where(date['days_name'].isin(['Sunday', 'Saturday']),1,0)
```

```python
date.head(10)
```

```python
date['date_week'] = date['date'].dt.week
```

```python
date.head()
```

```python
# Extract Quarter

date['quarter'] = date['date'].dt.quarter
```

```python
date.head(10)
```

```python
# Extract Semesterwise

date['half_yearly'] = np.where(date['quarter'].isin([1,2]),1,2)
```

```python
date.head(10)
```

# How to handle the mixed data

```python
df = pd.read_csv('titanic.csv')
df.head()
```

```python
df['number'].unique()
```

```python
df['number'].value_counts()
```

```python
fig = df['number'].value_counts().plot.bar()
fig.set_title("Passengers traelling")
```

```python
df['num_numerical']  = pd.to_numeric(df['number'], errors='coerce', downcast='integer')
```

```python
df.head()
```

```python
df['num_categorical'] = np.where(df['num_numerical'].isnull(), df['number'], np.nan)
```
